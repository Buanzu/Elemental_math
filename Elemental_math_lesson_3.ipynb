```python
import numpy as np
import matplotlib.pyplot as plt
round = np.linspace(0 , 2 * np.pi , 150)
radius = 0.4
a = radius * np.cos(round)
b = radius * np.sin(round)
figure, axes = plt.subplots( 1 )
axes.plot( a, b )
axes.set_aspect( 1 )
 
plt.show()
```

```python
# Задание № 5 график двух параллельных плоскостей. 
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-10, 10, 1.0)
y = np.arange(-10, 10, 1.0)
x, y = np.meshgrid(x,y)
Z1 = 2*x + 5*y
Z2 = 2*x + 5*y + 40
ax.plot_wireframe(x, y, Z1)
ax.plot_wireframe(x, y, Z2)
show()
```

```python
# Задание № 5 
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-10, 10, 1.0)
y = np.arange(-10, 10, 1.0)
x, y = np.meshgrid(x,y)
Z = x**2 + y**2
ax.plot_wireframe(x, y, Z)
show()
```
